```python
# Defining a function to calculate mean of Volume

l = []

def meanVolume(df):
    for x in df["Volume"]:
        l.append(x)
        
        mean = sum(l)/len(df)

    print(f"Mean of Volume is {mean:.2f}")
    
```

```python
# Calling the meanVolume function

meanVolume(NVDA_df)
```

```python
# Defining a function to calculte the trend of the NVDA closing stock price

def trend_identify(df, column = "Close"):
    df["Trend"] = df[column].diff().apply(lambda x: "Up" if x>0 else "Down" if x<0 else "Flat")
    return df
```

```python
trend_identify(NVDA_df)
```

```python
# Making a function for plotting a correlation heatmap

def correlation_close(df, parm = "Close"):
    new_df = df[["Volume", parm]]
    sns.heatmap(new_df.corr(), annot = True, cmap = "Greens")
```

# Section 4: Data Visualization (16 marks)  

```python
# Plotting Correlation heatmap for closing price and Volume


correlation_close(NVDA_df)
```

```python
# Scatter plot for plotting trends

import matplotlib.dates as mdates

# Mapping Trend to Numerical values:
trend_map = {'Up':1, 'Down':-1, 'Flat':0}
NVDA_df['TrendValue'] = NVDA_df['Trend'].map(trend_map)

# Plotting
plt.style.use('seaborn-v0_8-darkgrid')
fix, ax = plt.subplots(figsize=(12, 6))

# Plotting Close Prize Line:
ax.plot(NVDA_df['Date'], NVDA_df['Close'], label='Close Price', color='blue', alpha=0.8)

#Plotting Trend Scatter
scatter = ax.scatter(NVDA_df['Date'], NVDA_df['Close'],
                     c=NVDA_df['TrendValue'], cmap='RdYlGn',
                     label="Trend", s=10, alpha=0.6)

# Adding Color Bar:
cbar = plt.colorbar(scatter, ax=ax)
cbar.set_label('Trend (1: Up, -1: Down, 0: Flat)')

# Formatting X-axis for showing dates:
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=6))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.xticks(rotation=45) #Rotation for better readibility

plt.xlabel('Date')
plt.ylabel('Close Price')
plt.title('NVDA Stock Market Trends')
plt.legend()
plt.tight_layout()
plt.show
```
